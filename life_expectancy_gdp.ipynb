```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("all_data.csv")
data.columns = ["country", "year", "life_expectancy", "gdp"]
print(data.head())
```

```python
summary = data.describe()
print(summary)
```

```python
sns.lmplot(data=data, x="gdp", y="life_expectancy", hue="country", fit_reg=False)
plt.xlabel("Gross Domestic Product (USD)")
plt.ylabel("Life Expectancy (Age)")
plt.show()
plt.clf()
```

```python
sns.lmplot(data=data, x="gdp", y="life_expectancy", hue="year", fit_reg=False)
plt.xlabel("Gross Domestic Product (USD)")
plt.ylabel("Life Expectancy (Age)")
plt.show()
plt.clf
```

```python
sns.boxplot(data=data, x="country", y="life_expectancy")
plt.xlabel("Country")
plt.ylabel("Life Expectancy (Age)")
plt.xticks(rotation=30)
plt.subplot().set_xticks(range(6))
plt.subplot().set_xticklabels(["Chile", "China", "Germany", "Mexico", "USA", "Zimbabwe"])
plt.show()
plt.clf()
```

```python
life_expentancy_mean = data.groupby("country").mean()["life_expectancy"].reset_index()
life_expentancy_mean.sort_values(by="life_expectancy", inplace=True)
print(life_expentancy_mean, end="\n\n")

gdp_mean = data.groupby("country").mean()["gdp"].reset_index()
gdp_mean.sort_values(by="gdp", inplace=True)
print(gdp_mean)
```
